# Задание 1.
Будем парсить страницу со свежеми новостям на habr.com/ru/all/.

Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

KEYWORDS = ['python', 'парсинг']

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>



```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
KEYWORDS = ['python', 'парсинг']
```

```python
URL = 'https://habr.com/ru/search/'
params = {
    'q': KEYWORDS
}
```

```python
import time
time.sleep(0.2)
```

```python
def get_all_links(url, query):
    all_refs = []
    params = {
        'q': query,
    }
    res = requests.get(URL, params)
    time.sleep(0.2)
    soup = BeautifulSoup(res.text, 'html.parser')
    news_blocks = soup.find_all('li', class_='content-list__item content-list__item_post shortcuts_item')#блоки со статьями            
    articles_intro = list(map(lambda x: x.find('h2', class_='post__title'), news_blocks))#заголовки статей
    a_list = list(map(lambda x: x.find('a').get('href'), articles_intro))#ссылки на статьи
    return a_list

all_links = get_all_links(URL, KEYWORDS)
all_links
```

```python
import pandas as pd
```

```python
fresh_news = pd.DataFrame()
for link in all_links:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    date = soup.find('span', class_='post__time').text
    if soup.find('h2', class_='post__title'): 
        title = soup.find('h2', class_='post__title').text
    else:
        title = soup.find('h1',class_='post__title').text
    row = {'date': date, 'title': title, 'link': link}
    fresh_news = pd.concat([fresh_news, pd.DataFrame([row])])
```
